```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("/Advertising.csv")
df.head()
```

```python
df['total_spent']=df['TV']+df['radio']+df['newspaper']
df.head()
```

```python
sns.scatterplot
```

```python
sns.scatterplot(x='total_spent',y='sales',data=df)
```

```python
sns.regplot(x='total_spent',y='sales',data=df)
```

```python
x=df.drop(['sales','total_spent'],axis=1)
y=df['sales']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.preprocessing import PolynomialFeatures
plo=PolynomialFeatures(degree=4,include_bias=False)
```

```python
pp=plo.fit_transform(x)
```

```python
pp.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(pp,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
mm=model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
mae=mean_absolute_error(y_test,mm)
mse=mean_squared_error(y_test,mm)
rmse=np.sqrt(mse)
aa=r2_score(y_test,mm)
```
